## Data Cleaning
The product filter types that will be used for my recommendation system was extracted as a paragraph and sits within the same cell. (An example can be seen in cell[4] below). Hence, I wrote a regex function to split each filter type into individual columns. These columns will go through pre-processing in the next notebook.

```python
# filter types in the same cell with '\n' acting as a separator filter type
cleanser_raw['desc_summary'][0]
```

```python
# Regex function to split filter types in 'desc_summary' into individal columns
def split_desc_summary(df):
    
    # add '\n' to end of string to support regex search pattern
    df['desc_summary'] = [i+'\n' for i in df['desc_summary']]
    
    # split features in 'desc_summary' into individual columns
    skin_type_col = []
    skin_concerns_col = []
    formulation_col = []
    age_col = []
    
    for i in range(len(df)):
        desc = df.iloc[i]['desc_summary']

        try:
            skintype = re.search("(?<=Skin Type: ).+?(?=\\n)", desc)[0]
            skin_type_col.append(skintype)
        except: 
            skintype = 'NA'
            skin_type_col.append(skintype)
        
        try:
            skinconcerns = re.search("(?<=Skin Concerns: ).+?(?=\\n)", desc)[0]
            skin_concerns_col.append(skinconcerns)
        except: 
            skinconcerns = 'NA'
            skin_concerns_col.append(skinconcerns)
            
        try:    
            formulation = re.search("(?<=Formulation: ).+?(?=\\n)", desc)[0]
            formulation_col.append(formulation)
        except: 
            formulation = 'NA'
            formulation_col.append(formulation)
        
        try:
            age = re.search("(?<=Skincare By Age: ).+?(?=\\n)", desc)[0]
            age_col.append(age)
        except:
            age = 'NA'
            age_col.append(age)
    
    # append new columns to dataset
    df['skin_type'] = skin_type_col
    df['skin_concerns'] = skin_concerns_col
    df['formulation'] = formulation_col
    df['age'] = age_col
    
    df.drop(columns = ['desc_summary', 'what_it_is', 'what_it_does', 'pdt_claims', 'pdt_ingredients'], inplace = True) # drop unnecessary column   
    return df
```

```python
# split filter types in `desc summary`
cleanser = split_desc_summary(cleanser_raw)
toner = split_desc_summary(toner_raw)
day_moisturizer = split_desc_summary(day_moisturizer_raw)
night_cream = split_desc_summary(night_cream_raw)
sunscreen = split_desc_summary(sunscreen_raw)
```

```python
# Creating variables to be used in for loops below
categories_name = ['Cleanser', 'Toner', 'Day Moisturizer', 'Night Cream', 'Sunscreen']
categories = [cleanser, toner, day_moisturizer, night_cream, sunscreen]
len(categories)
```

```python
# as ratings follow a left skewed distribution, I will impute ratings with null value with the median 
##(Graph can be found below, cell[42])
# impute missing number of reviews, product claims and product ingredients with 'NA'
for i in categories:
    rating_median = i['rating'].median()
    i['rating'].replace(np.nan, rating_median, inplace = True)
    i['num_reviews'].replace(np.nan, 0, inplace = True)
```

```python
# Add column to label products by their category
labels = ['cleaser', 'toner', 'day_moisturizer', 'night_cream', 'sunscreen']
for m in range(5):       
    categories[m]['category'] = labels[m]
```

```python
# Map sephora exclusive products as 1 and non-exclusive products as 0.
## A Sephora Exclusive product means the product only Sephora carries the product and it cannot be found in other stores
for i in categories:
    i['sephora_exclusive'] = i['sephora_exclusive'].map({'SEPHORA EXCLUSIVE': 1, 'NEW | SEPHORA EXCLUSIVE':1, '0':0})
```

```python
# Remove unnecessary spaces to prepare for one-hot encoding
for i in categories:
    i['skin_type'] = i['skin_type'].str.replace(' ', '')
    i['skin_concerns'] = i['skin_concerns'].str.replace(' ', '')
    i['formulation'] = i['formulation'].str.replace(' ', '')
    i['age'] = i['age'].str.replace(' ', '')
```

```python
# Change data type for price to integer
for i in categories:
    i['price'] = i['price'].astype(int)
```

```python
# Check for any more null values in the datasets
for l in range(5):
    print(f'Null values in {categories_name[l]}: {categories[l].isnull().sum().sum()}')
```

```python
# Drop brand only for men
for i in categories: 
    drop_jb = i[i['brand'] == 'JACK BLACK'].index
    i.drop(drop_jb, inplace = True)
```

```python
# function for plotting historgrams of skincare features

def subplot_hist(col, figtitle, xlabel):
    fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (16,8))
    ax= ax.ravel()

    cl_plt = sns.histplot(cleanser[col], ax = ax[0], color = '#1BAFD0')
    to_plt = sns.histplot(toner[col], ax = ax[1], color = '#FD636B')
    dm_plt = sns.histplot(day_moisturizer[col], ax = ax[2], color = '#FFB900')
    nc_plt = sns.histplot(night_cream[col], ax = ax[3], color = '#3BE8B0')
    ss_plt = sns.histplot(sunscreen[col], ax = ax[4], color = '#6967CE')
    plt.tight_layout(h_pad = 5)
    
    cl_plt.axvline(x = cleanser[col].mean(), c='k', ls = '--')
    to_plt.axvline(x = toner[col].mean(), c='k', ls = '--')
    dm_plt.axvline(x = day_moisturizer[col].mean(), c='k', ls = '--')
    nc_plt.axvline(x = night_cream[col].mean(), c='k', ls = '--')
    ss_plt.axvline(x = sunscreen[col].mean(), c='k', ls = '--')
    
    plt.suptitle(figtitle, size = 18, y = 1.1)
    cl_plt.set_title(f'Cleanser\nMean:{round(cleanser[col].mean(),2)}', size = 14)
    to_plt.set_title(f'Toner\nMean:{round(toner[col].mean(),2)}', size = 14)
    dm_plt.set_title(f'Day Moisturizer\nMean:{round(day_moisturizer[col].mean(),2)}', size = 14)
    nc_plt.set_title(f'Night Cream\nMean:{round(night_cream[col].mean(),2)}', size = 14)
    ss_plt.set_title(f'Sunscreen\nMean:{round(sunscreen[col].mean(),2)}', size = 14)

    for i in [cl_plt, to_plt, dm_plt, nc_plt, ss_plt]:
        i.set_xlabel(xlabel, size = 12)

    for j in [cl_plt, to_plt, dm_plt, nc_plt, ss_plt]:
        j.set_ylabel('Count', size = 12)

    ax[-1].set_visible(False) # hide unused plots
```

```python
subplot_hist('rating', 'Histogram of Ratings for Each Category', 'Product Ratings')
```
